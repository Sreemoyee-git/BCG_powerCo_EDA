```python
plt.figure(figsize=(10, 4))
sns.histplot(Client_df['cons_12m'], bins=50)
plt.title('12-Month Electricity Consumption Distribution')
plt.xlabel('cons_12m')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sample mapping for channel_sales
channel_sales_map = {
    "foosdfpfkusacimwkcsosbicdxkicaua": "Online Direct",
    "lmkebamcaaclubfxadlmueccxoimlema": "Tele Sales",
    "usilxuppasemubllopkaafesmlibmsdf": "Email Campaign",
    "ewpakwlliwisiwduibdlfmalxowmwpci": "Partner Referral",
    "epumfxlbckeskwekxbiuasklxalciiuu": "Physical Store",
    "fixdbufsefwooaasfcxdxadsiekoceaa": "Mobile App",
    "sddiedcslfslkfkowfkldpoeeailfpeds": "Social Media",
    "MISSING": "Unknown Channel"
}

# Sample mapping for origin_up
origin_up_map = {
    "lxidpiddsbxsbosboudacockeimpuepw": "Website Signup",
    "kamkkxfxxuwbdslkwifmmcsiusiuosws": "In-Store Signup",
    "ldkssxwpmemidmecebumciepifcamkci": "Phone Signup",
    "usapbepcfoloekilkwsdiboslwaxobdp": "Referral",
    "ewxeelcelemmiwuafmddpobolfuxioce": "Email Campaign",
    "MISSING": "Unknown Source"
}

# Apply mappings to your DataFrame
Client_df["channel_sales_mapped"] = Client_df["channel_sales"].map(channel_sales_map)
Client_df["origin_up_mapped"] = Client_df["origin_up"].map(origin_up_map)

# Now plot the cleaned versions
plt.figure(figsize=(8, 4))
sns.countplot(y='channel_sales_mapped', data=Client_df, order=Client_df['channel_sales_mapped'].value_counts().index)
plt.title("Channel Sales Distribution")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
sns.countplot(y='origin_up_mapped', data=Client_df, order=Client_df['origin_up_mapped'].value_counts().index)
plt.title("Origin Up Distribution")
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='churn', data=Client_df)
plt.title("Churn Class Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```

```python
price_agg = Price_df.groupby("id").mean(numeric_only=True).reset_index()
merged_df = pd.merge(Client_df, price_agg, on="id", how="left")
```

```python
correlation = merged_df.corr(numeric_only=True)
sns.heatmap(correlation[['churn']].sort_values(by='churn', ascending=False), annot=True, cmap="coolwarm")
plt.title("Feature Correlation with Churn")
plt.show()
```
